```python
%reset -f
import utils.TimeDomainAnalysis as TD

import pandas as pd
import numpy as np

PATH = './Data/Case/48k/'
```

```python
# 时域分析
fig1 = pd.read_csv(PATH + 'csvFigure/N_FE.csv')
fig2 = pd.read_csv(PATH + 'csvFigure/N_DE.csv')
fs = 48000

# 概率密度分布
PDF = TD.ProbabilityDensity(fig1, 50)
# print(PDF.Hist)
PDF.plot()
# 近似为正态分布
```

```python
# 自相关与互相关函数
signal1 = np.squeeze(np.array(fig1))
signal1 = signal1[:960]
signal2 = np.squeeze(np.array(fig2))
signal2 = signal2[:960]

Auto = TD.Correlation(fs, signal1)
Auto.plot()
Auto = TD.Correlation(fs, signal2)
Auto.plot()
# 可见不同信号的自相关函数形态相差较大，是一个不错的分类指标
Cross = TD.Correlation(fs, signal1, signal2)
Cross.plot()
```
